```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Create the dataset
data = {
    'ID': [1, 2, 3, 4, 5],
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve'],
    'Age': [30, 25, 28, 35, 40],
    'Salary': [70000, None, 60000, None, 80000],
    'Department': ['HR', 'IT', 'Marketing', 'IT', 'Finance']
}

df = pd.DataFrame(data)

# Fill missing salary values with the mean
df['Salary'] = df['Salary'].fillna(df['Salary'].mean())

# Set the visual style of seaborn
sns.set(style="whitegrid")

# 1. Bar Plot: Average Salary by Department
plt.figure(figsize=(10, 6))
sns.barplot(x='Department', y='Salary', data=df, errorbar=None)
plt.title('Average Salary by Department')
plt.ylabel('Average Salary')
plt.xlabel('Department')
plt.show()

# 2. Box Plot: Salary Distribution by Department
plt.figure(figsize=(10, 6))
sns.boxplot(x='Department', y='Salary', data=df)
plt.title('Salary Distribution by Department')
plt.ylabel('Salary')
plt.xlabel('Department')
plt.show()

# 3. Histogram: Salary Distribution
plt.figure(figsize=(10, 6))
sns.histplot(df['Salary'], bins=5, kde=True)
plt.title('Salary Distribution')
plt.xlabel('Salary')
plt.ylabel('Frequency')
plt.show()

# 4. Scatter Plot: Age vs. Salary
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Age', y='Salary', hue='Department', style='Department', data=df, s=100)
plt.title('Age vs. Salary')
plt.xlabel('Age')
plt.ylabel('Salary')
plt.legend(title='Department')
plt.show()

# 5. Count Plot: Department Distribution
plt.figure(figsize=(10, 6))
sns.countplot(x='Department', data=df)
plt.title('Number of Employees by Department')
plt.xlabel('Department')
plt.ylabel('Number of Employees')
plt.show()
```
